### Problem 2
The factorial of any positive integer $k$ is defined as:
$$k! = k \times (k-1) \times \cdots \times 2 \times 1$$
Compute the following expression:
$$\sum_{k=1}^9 k!$$

```python
def kfact(k):                      # Create a function that calculates factorial
    if k == 1:                     # Factorial of 1 is 1. We do not need to be concerned about zero since the question specifies 1 to 9.
        return 1
    else:
        return (k * kfact(k - 1))  # When k is not one, the factorial is k multiplied by the factorial of k - 1, hence the recursive formula

for i in range(1, 10):             # Define a range of integers that includes 1 and 9 
    factorial = kfact(i)           # Apply the kfact function to the integer
    print(factorial)               # Print the factorial
    i += 1                         # Do the last 2 steps for all i in the range.
```

### Problem 3
Give a string consists of several letters. If the letter is in the even-position, convert it to lower case; if the letter is in the odd-position, convert it to upper case.
For example, if the given string is "Morning", the result should be "MoRnInG".

Convert the string "Welcome" and print the result.

```python
str = list('Welcome')             # Define the string to be worked with in a list so all characters can be handled separately
for i in range(1, len(str)):      # Define a range that starts from 1 to the length of the string (e.g. Welcome has 7 letters, 
                                  # so len(Welcome) = 7
    str[0] = str[0].upper()       # Since Python reads index from 0, the 0th position is typically an odd position (position 1) 
                                  # so it will always be in upper case. This eventually causes the next position to be even, then, odd, etc.
    if i % 2 == 1:                # This tells that we have an odd index (but now even)
        str[i] =  str[i].lower()  # So convert the letter to a lowercase.
    else:
        str[i] =  str[i].upper()  # Otherwise, convert to an uppercase.
print(''.join(str))               # Print the new string after unlisting the characters.
```

### Problem 5
In number theory, the Padovan sequence is the sequence of integers $P(n)$ defined by the initial values
$$P(0) = P(1) = P(2) = 1$$
and the recurrence relation
$$P(n) = P(n-2) + P(n-3)$$
For example, the first few values of $P(n)$ are
1, 1, 1, 2, 2, 3, 4, 5, 7, 9

Compute and print $P(20)$.nt 
.

```python
def P(n):                                           # Define a function that calculates the Padovan sequence
    if (n < 0 or isinstance(n, int) == False):      # If n is not positive or n is not an integer, raise an error.
        raise Exception('The input is either a negative number, a non-integer or a character')
    elif (n == 0 or n == 1 or n == 2):              # Define the initial values
        P_n = 1
        return(P_n)
    else:
        P_n = P(n - 2) + P(n - 3)                   # Outside the initial values, calculate Padovan sequence of n as a recurrence relation. 
        return(P_n)                                 # Return the answer
print(P(20))                                        # Insert any n in P(n) and print the Padovan sequence
```
