### Problem 1
Please use the math formulas in the lecture notes to compute the solution for the linear regression:
$$
\hat{\mathbf{w}} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}
$$
You can use Numpy package, but cannot use scikit-learn package.

```python
import numpy as np

def manual_lin_reg(X, y):
    w = np.dot((np.linalg.inv(np.dot(np.transpose(X), X))), np.dot(np.transpose(X), y)) # This is the formula for the coefficient w.
    return(w)

X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]]) # This defines the variables X
col_of_ones = np.ones((len(X), 1)) 
X = np.append(X, col_of_ones, axis = 1) # This adds ones as the last column of X
# print(X)

y = np.dot(X, np.array([1, 2, 0])) + 3 # # This defines the variable y
# print(y)

w = manual_lin_reg(X, y) # This applies the manual regression calculation function to dataset (X, y)
print("w = ", w)
```

In the section for linear regression with multiple explanatory variables, prove that $\textbf{X}^T\textbf{X}$ is positive definite if and only if $\textbf{X}$ has full column rank.

### Proof

Let $ X^TX $ be positive definite. \
Assume X does not have a full column rank. \
$ \implies $ The columns of $ X $ are not linearly independent. \
$ \implies \exists $ a column vector $ v \neq 0 \backepsilon $ 
$$ Xv = 0 $$
So $$ v^T X^T X v = 0 $$ 
Since $ X^TX $ is positive definite, $ v^T X^T X v = 0 \implies v = 0 $
$$ \rightarrow\!\leftarrow $$
$ \therefore X $ has full column rank.

Conversely, \
Let $ X $ have full column rank. \
Then all columns of $ X $ are linearly independent. \
$ \implies \forall $ non-zero column vectors $ v, \,  v^TXv = 0 $ when  $ v = 0 $ and $ v^TXv > 0 $ when $ v > 0 $. \
$ \implies v^T X^T X v \geq 0 $ \
$ \therefore X^T X $ is positive definite.
